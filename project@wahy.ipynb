```python
#Extracting with the total numbers
total=data[(data["DISTRICT"]=="TOTAL")]
total
```

```python
ker_dist=data[(data["STATE/UT"]=="KERALA")]
ker_dist
```

```python
#creating data fram of deatis of kerala state
kerala=total[(total["STATE/UT"]=="KERALA")]
```

```python
import matplotlib.pyplot as plt
for i in kerala:
    sns.barplot(x=kerala["YEAR"],y=kerala[i])
    plt.show()
```

```python
import matplotlib.pyplot as plt
for i in kerala:
    sns.lineplot(x=kerala["YEAR"],y=kerala[i])
    plt.show()
```

```python
#grouping
data_murder=pd.DataFrame(total.groupby(["STATE/UT"])["MURDER"].sum().reset_index())
data_murder
data_murder.sort_values("MURDER",ascending=False)
```

```python
data_crimes=pd.DataFrame(total.groupby(["STATE/UT"])["TOTAL IPC CRIMES"].sum().reset_index())
data_crimes
data_crimes.sort_values("TOTAL IPC CRIMES",ascending=False)
```
